```python

import pandas as pd
import numpy as np
import scipy.stats as st
import plotly.graph_objs as go
import pandas as pd


global dados

def puxar_ipca(data_inicio, data_fim, fig3):
  # Definindo o código do IPCA no SGS
  codigo_ipca = 433

  # Obtendo dados do IPCA mensal
  ipca_mensal = sgs.get({'IPCA': codigo_ipca}, start=data_inicio)


  # Calculando o acumulado dos últimos 12 meses
  ipca_mensal['IPCA Acumulado 12 Meses'] = ipca_mensal['IPCA'].rolling(window=12).sum()


  print(ipca_mensal.dropna(subset=['IPCA Acumulado 12 Meses']))

  # Visualização com plotly
  fig = go.Figure()

  # Adicionando o IPCA mensal

  fig3.add_trace(go.Scatter(x=ipca_mensal.index, y=ipca_mensal['IPCA'], mode='lines+markers', name='IPCA Mensal', line=dict(color='Blue',dash='dash', width=2)))


  # Adicionando o IPCA acumulado nos últimos 12 meses
  fig.add_trace(go.Scatter(x=ipca_mensal.index, y=ipca_mensal['IPCA Acumulado 12 Meses'], mode='lines+markers', name='Acumulado 12 Meses'))

  fig.update_layout(title='IPCA Mensal e Acumulado',
                  xaxis_title='Data',
                  yaxis_title='IPCA')

  fig.show()

  return




def calcular_valor_ntnb_com_grafico(valor_nominal, cupom, ipca_acumulado, anos_ate_vencimento, taxa_mercado):

    valor_ajustado_ipca = valor_nominal * (1 + ipca_acumulado / 100)
    juros_anuais = valor_ajustado_ipca * (cupom / 100)
    juros_mensais = juros_anuais / 12

    valor_presente = 0
    valores_mensais = []
    juros_mensais_lista = []

    meses_ate_vencimento = anos_ate_vencimento * 12

    for mes in range(1, meses_ate_vencimento + 1):
        valor_mensal = juros_mensais / ((1 + taxa_mercado / 100 / 12) ** mes)
        valores_mensais.append(valor_mensal)
        juros_mensais_lista.append(juros_mensais)
        valor_presente += valor_mensal

    valor_presente += valor_ajustado_ipca / ((1 + taxa_mercado / 100 / 12) ** meses_ate_vencimento)

    # Plotando o gráfico
    fig = go.Figure()

    fig.add_trace(go.Scatter(x=list(range(1, meses_ate_vencimento + 1)), y=juros_mensais_lista, mode='lines', name='Juros Mensais'))
    fig.add_trace(go.Scatter(x=list(range(1, meses_ate_vencimento + 1)), y=valores_mensais, mode='lines', name='Valor Presente Mensal'))
    fig.add_trace(go.Scatter(x=[meses_ate_vencimento], y=[valor_ajustado_ipca / ((1 + taxa_mercado / 100 / 12) ** meses_ate_vencimento)], mode='markers', name='Valor Nominal Ajustado'))

    fig.update_layout(
        title='Valores do Título NTN-B ao Longo do Tempo (Mensal)',
        xaxis_title='Meses',
        yaxis_title='Valor (R$)',
        showlegend=True
    )

    fig.show()

    return



import pandas as pd
import plotly.graph_objects as go

def separar_volume():
    print("separar_volume")
      # Pass 'dados' as a parameter instead of using 'global'

    # Seleciona as colunas necessárias
    volume = dados[['CAPTC_DIA', 'RESG_DIA', 'VL_QUOTA', 'CNPJ_FUNDO_CLASSE']]

    # Chama a função para criar o gráfico de volume
    volume = go.Scatter(
        x=volume.index,
        y=volume['VL_QUOTA'],
        mode='lines+markers',
        line=dict(dash='dash')
    )

    fig = go.Figure(data=volume)
    fig.update_layout(
        title='Volume Chart',
        xaxis_title='Date',
        yaxis_title='Volume (VL_QUOTA)',
        template='plotly_dark'
    )


def separar_fundo(cnpjs):
    print("separar_fundo")

    dados['DT_COMPTC'] = pd.to_datetime(dados['DT_COMPTC'], format='%Y-%m-%d')
    fundos = dados[dados['CNPJ_FUNDO_CLASSE'].isin(cnpjs)].sort_values('DT_COMPTC')[['DT_COMPTC', 'CNPJ_FUNDO_CLASSE', 'VL_QUOTA']]
    fundos.to_excel('cnpj_separado.xlsx', index=False)
    print(fundos.head())

    return fundos

def volume(cnpjs):
    print("volume")
    global dados
    volume = dados[dados['CNPJ_FUNDO_CLASSE'].isin(cnpjs)].sort_values('DT_COMPTC')[['CAPTC_DIA', 'RESG_DIA', 'VL_QUOTA', 'CNPJ_FUNDO_CLASSE']]
    print(f'Volume \n\n {volume.head()}')
    return volume

def calcular_sharpe(fundos, taxa_cdi_diaria=0.00049037	):
    print("calcular_sharpe")
    global df
    df = fundos.copy()
    df.set_index('DT_COMPTC', inplace=True)
    df['Retorno Diário'] = np.log(df['VL_QUOTA'] / df['VL_QUOTA'].shift(1))
    df.dropna(subset=['Retorno Diário'], inplace=True)

    retorno_medio = df['Retorno Diário'].mean()
    desvio_padrao = df['Retorno Diário'].std()
    indice_sharpe = (retorno_medio - taxa_cdi_diaria) / desvio_padrao
    print(f'#################################### \n Índice de Sharpe da série: {indice_sharpe:.4f}\n ##################')

    Media_curta = 5
    Media_longa = 10
    df['Média Móvel Longa dias'] = df['VL_QUOTA'].rolling(window=Media_longa).mean()
    df['Média Móvel curta dias'] = df['VL_QUOTA'].rolling(window=Media_curta).mean()

    rentabilidade(df)

    return df

def grafico(cnpjs):
    print("grafico")
    global df
    fig = go.Figure()

    fig.add_trace(go.Scatter(x=df.index, y=df['VL_QUOTA'], mode='lines', name='Série Original'))
    fig.add_trace(go.Scatter(x=df.index, y=df['Média Móvel Longa dias'],  mode='lines', line=dict(dash='dash'), name='Média Móvel Longa dias'))
    fig.add_trace(go.Scatter(x=df.index, y=df['Média Móvel curta dias'], mode='lines', name='Média Móvel curta dias'))

    fig.update_layout(
        title=f'Série Original e Médias Móveis - {cnpjs}',
        xaxis_title='Data',
        yaxis_title='Valores'
    )

    fig.show()

def var():
    print("var")
    global df
    confidence_level = 0.95
    Var_curto = 20
    Var_longo = 60

    var = st.norm.ppf(1 - confidence_level, loc=df['Retorno Diário'].mean(), scale=df['Retorno Diário'].std()) * np.sqrt(Var_curto)
    mensagem = f"Valor de Risco (VaR) para {Var_longo} dias, com {confidence_level*100}% de confiança: {var}"
    print(Fore.RED + mensagem + Style.RESET_ALL)
    var_10dias = st.norm.ppf(1 - confidence_level, loc=df['Retorno Diário'].mean(), scale=df['Retorno Diário'].std()) * np.sqrt(Var_longo)
    mensagem = f"Valor de Risco (VaR) para {Var_curto} dias, com {confidence_level*100}% de confiança: {var_10dias}"
    print(Fore.RED + mensagem + Style.RESET_ALL)

    return

def rentabilidade(df):
    print("rentabilidade")
    global base_final
    rentabilidade = pd.DataFrame()
    rentabilidade['Rentabilidade Diária'] = df['VL_QUOTA'].pct_change()
    rentabilidade['Rentabilidade Acumulada'] = (1 + rentabilidade['Rentabilidade Diária']).cumprod() - 1

    data_inicio = min(df.index)
    data_fim = max(df.index)
    # base_final["CDI_Acumulado"] = puxar_cdi(data_inicio, data_fim)
    # rentabilidade['Rentabilidade Acumulada CDI'] = base_final['CDI_Acumulado']

    fig3 = go.Figure()
    fig3.add_trace(go.Scatter(x=df.index, y=rentabilidade['Rentabilidade Acumulada'], mode='lines', name='Rentabilidade Acumulada', line=dict(color='blue', width=2)))
    # fig3.add_trace(go.Scatter(x=df.index, y=rentabilidade['Rentabilidade Acumulada CDI'], mode='lines', name='Rentabilidade Acumulada CDI', line=dict(color='red', width=2)))
    fig3.update_layout(
        title='Rentabilidade do Fundo',
        xaxis_title='Data',
        yaxis_title='Valores'
    )

    # Solicitar confirmação do usuário
    resposta = input("Deseja inserir os dados do dólar? Sim/Não: ").strip().lower()

    if resposta == "sim":
        puxar_dolar(data_inicio, data_fim, fig3)
    else:
        print('Continuar')

     # Solicitar confirmação do usuário
    resposta = input("Deseja inserir os dados do CDI? Sim/Não: ").strip().lower()

    if resposta == "sim":
        puxar_cdi(data_inicio, data_fim, fig3)
    else:
        print('Continuar')

    resposta = input("Deseja inserir os dados do IPCA? Sim/Não: ").strip().lower()
    if resposta == "sim":
        puxar_cdi(data_inicio, data_fim, fig3)
    else:
        print('Continuar')


    fig3.show()

def puxar_dolar(data_inicio, data_fim, fig3):
    print("puxar_dolar")

    codigo_dolar = 1

    # Obtém os dados do dólar comercial usando a função 'get' da biblioteca 'sgs'
    dolar_diario = sgs.get([codigo_dolar], start=data_inicio, end=data_fim)

    # Renomeia a coluna para melhor legibilidade
    dolar_diario.columns = ['Dolar_Fechamento']
    dolar_diario['Dolar_Rentabilidade'] = dolar_diario['Dolar_Fechamento'].pct_change()
    dolar_diario['Dolar_Fechamento Acumulada'] = (1 + dolar_diario['Dolar_Rentabilidade']).cumprod() - 1

    fig3.add_trace(go.Scatter(x=dolar_diario.index, y=dolar_diario['Dolar_Fechamento Acumulada'], mode='lines', name='Dolar Acumulado', line=dict(color='red',dash='dash', width=2)))

    # Exibe os dados do dólar
    print("A seguir a tabela")
    print(dolar_diario.head())
    print("fim")



def puxar_cdi(data_inicio,data_fim,fig3):
  print("Puxar CDI")
  # Define o código da série do CDI diário (série 12)
  codigo_cdi = 12

  # Define o período para obter os dados (substitua pelas datas desejadas)


  # Obtém os dados do CDI diário usando a função 'get' da biblioteca 'sgs'
  cdi_diario = sgs.get({'cdi': codigo_cdi}, start=data_inicio, end=data_fim)
  print(cdi_diario)
  # Renomeia a coluna 'cdi' para 'CDI' para melhor legibilidade
  cdi_diario = cdi_diario.rename(columns={'cdi': 'CDI'})
  Dados_cdi=pd.DataFrame()
  Dados_cdi['CDI']=(cdi_diario['CDI'])
  Dados_cdi['CDI_Fechamento Acumulada'] = (1 + (cdi_diario['CDI']/100)).cumprod() - 1

  fig3.add_trace(go.Scatter(x=Dados_cdi.index, y=Dados_cdi['CDI_Fechamento Acumulada'], mode='lines', name='CDi Acumulado', line=dict(color='green',dash='dashdot', width=2)))

  return
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objects as go

# Processamento de dados como antes...
dados['DT_COMPTC'] = pd.to_datetime(dados['DT_COMPTC'], format='%Y-%m-%d')
cnpj_Cor = ["38.236.242/0001-51", "28.578.897/0001-54"]
separar_fundo(cnpj_Cor)
dados_correlacao = dados[dados['CNPJ_FUNDO_CLASSE'].isin(cnpj_Cor)].sort_values('DT_COMPTC')[['DT_COMPTC', 'CNPJ_FUNDO_CLASSE', 'VL_QUOTA']]
dados_agrupados = dados_correlacao.groupby(['DT_COMPTC', 'CNPJ_FUNDO_CLASSE']).mean().reset_index()
nomes_fundos = {
    "38.236.242/0001-51": "BB europeia BDR",
    "28.578.897/0001-54": "Caixa BDR"
}
dados_pivot = dados_agrupados.pivot(index='DT_COMPTC', columns='CNPJ_FUNDO_CLASSE', values='VL_QUOTA')
dados_pivot.rename(columns=nomes_fundos, inplace=True)
dados_pivot.to_excel('fundos.xlsx', index=True)
correlation_matrix = dados_pivot.corr()
plt.figure(figsize=(12, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, cbar_kws={'label': 'Correlação'})
plt.title('Matriz de Correlação')
plt.show()
dados_pivot['data'] = dados_agrupados['DT_COMPTC']

# Função da fronteira eficiente
def fronteira_eficiente(retornos, num_portfolios=10000):
    num_ativos = len(retornos.columns)
    retornos_portfolio = np.zeros(num_portfolios)
    volatilidades_portfolio = np.zeros(num_portfolios)
    sharpe_ratios = np.zeros(num_portfolios)
    pesos_portifolio = []
    retornos = retornos.apply(pd.to_numeric, errors='coerce').dropna()
    retornos_medios = retornos.mean()
    matriz_cov = retornos.cov()
    taxa_livre_risco = 0
    for i in range(num_portfolios):
        pesos = np.random.random(num_ativos)
        pesos /= np.sum(pesos)
        pesos_portifolio.append(pesos)
        retornos_portfolio[i] = np.dot(pesos, retornos_medios)
        volatilidades_portfolio[i] = np.sqrt(np.dot(pesos.T, np.dot(matriz_cov, pesos)))
        sharpe_ratios[i] = (retornos_portfolio[i] - taxa_livre_risco) / volatilidades_portfolio[i]
    return retornos_portfolio, volatilidades_portfolio, sharpe_ratios, np.array(pesos_portifolio)

# Exemplo de uso com dados fictícios
retornos = pd.DataFrame(dados_pivot)
retornos['data'] = pd.to_datetime(retornos['data'])
retornos.set_index('data', inplace=True)
retornos_portfolio, volatilidades_portfolio, sharpe_ratios, pesos_portifolio = fronteira_eficiente(retornos)
indice_max_sharpe = sharpe_ratios.argmax()
max_sharpe_return = retornos_portfolio[indice_max_sharpe]
print(f'max_sharpe_return {max_sharpe_return}')
max_sharpe_volatility = (volatilidades_portfolio[indice_max_sharpe]/100)
print(f'max_sharpe_return {max_sharpe_volatility}')

print(f'Retorno do portfólio com o maior índice de Sharpe: {max_sharpe_return}')
print(f'Volatilidade do portfólio com o maior índice de Sharpe: {max_sharpe_volatility}')
max_sharpe_portfolio = pd.DataFrame(pesos_portifolio[indice_max_sharpe], index=retornos.columns, columns=['Peso'])
print(max_sharpe_portfolio.head())

# Plot da fronteira eficiente com Plotly
fig = go.Figure()

# Adicionando a dispersão dos portfólios
fig.add_trace(go.Scatter(
    x=volatilidades_portfolio,
    y=retornos_portfolio,
    mode='markers',
    marker=dict(
        color=sharpe_ratios,
        colorscale='Viridis',
        colorbar=dict(title='Índice de Sharpe'),
        showscale=True
    ),
    text=[f'Índice de Sharpe: {sr:.2f}' for sr in sharpe_ratios],
    name='Portfólios'
))

# Adicionando o portfólio com o maior índice de Sharpe
fig.add_trace(go.Scatter(
    x=[max_sharpe_volatility],
    y=[max_sharpe_return],
    mode='markers',
    marker=dict(
        color='red',
        size=12,
        symbol='star'
    ),
    text=['Máx de Sharpe'],
    name='Máx de Sharpe'
))

# Ajustando layout do gráfico
fig.update_layout(
    title='Fronteira Eficiente com Índice de Sharpe',
    xaxis_title='Volatilidade',
    yaxis_title='Retorno',
    legend=dict(x=0.8, y=0.2)
)

# Exibindo o gráfico
fig.show()

# Criação do gráfico de pizza para a composição do portfólio com maior índice de Sharpe
fig2 = go.Figure(data=[go.Pie(labels=max_sharpe_portfolio.index, values=max_sharpe_portfolio['Peso'])])
fig2.update_layout(title='Composição do Portfólio com o Maior Índice de Sharpe')

# Exibindo o gráfico de pizza
fig2.show()
```

```python
###### FOCUS  BACEN parte 2


# Import the necessary libraries
from pickle import NONE
from bcb import Expectativas
import pandas as pd
import plotly.graph_objs as go
from plotly.subplots import make_subplots
import numpy as np

# Instancia a classe
em = Expectativas()


while True:  # Loop para garantir que data_inicio seja definida
    try:
        # Verifica se data_inicio já foi definida em outro lugar
        if 'data_inicio' not in locals() or data_inicio is None or data_inicio == 0:
            data_inicio = '2024-01-01'  # Define um valor padrão para data_inicio
            print(f"Data início foi definida como: {data_inicio}")
        break  # Sai do loop se data_inicio já foi corretamente inicializada
    except Exception as e:
        print(f"Ocorreu um erro: {e}")
        data_inicio = None  # Redefine a variável e tenta novamente



# Obtém os entitysets (nome das pesquisas)
#print(em.describe())

# Obtém as informações da pesquisa
print(em.describe('ExpectativasMercadoSelic'))

# Conectar com a API do Banco Central e executar a consulta
ipca_expec_mensais = em.get_endpoint('ExpectativaMercadoMensais')
ipca_expec_ano = em.get_endpoint('ExpectativasMercadoAnuais')
SELIC_expec_mensais = em.get_endpoint('ExpectativasMercadoSelic')

# Dados do IPCA (mensais)
# Check available data range for Indicador 'IPCA'
try:
  print("ipca Mensal")
  ipca_expec_mensais = (
      ipca_expec_mensais.query()
      .filter(ipca_expec_mensais.Indicador == 'IPCA')
      .filter(ipca_expec_mensais.Data >= '2024-06-01')
      .select(ipca_expec_mensais.Indicador, ipca_expec_mensais.Data, ipca_expec_mensais.Media, ipca_expec_mensais.Mediana, ipca_expec_mensais.DataReferencia)
      .collect()
  )
except:
  print("ipca mensal falhou")
try:
  # Check available data range for Indicador 'SELIC'
  SELIC_expec_mensais = (
    SELIC_expec_mensais.query()
    .filter(SELIC_expec_mensais.Indicador == 'Selic')
    .filter(SELIC_expec_mensais.Data >= '2024-06-01')
    .select(SELIC_expec_mensais.Indicador, SELIC_expec_mensais.Data, SELIC_expec_mensais.Media, SELIC_expec_mensais.Mediana, SELIC_expec_mensais.Reuniao)
    .collect()
  )
except:
  print("Selic Falhou")
try:
  # Dados do IPCA (anuais)
  print("ipca anual")
  ipca_expec_ano = (ipca_expec_ano.query()
      .filter(ipca_expec_ano.Indicador == 'IPCA', ipca_expec_ano.DataReferencia == 2025)
      .filter(ipca_expec_ano.Data >= '2024-01-01')
      .filter(ipca_expec_ano.baseCalculo == '0')
      .select(ipca_expec_ano.Indicador, ipca_expec_ano.Data, ipca_expec_ano.Media, ipca_expec_ano.Mediana, ipca_expec_ano.DataReferencia)
      .collect()
  )
except:
  print("ipca anual falhou")

# Verify the columns in the DataFrame


# Formata a coluna de Data para formato datetime
ipca_expec_mensais['Data'] = pd.to_datetime(ipca_expec_mensais['Data'], format='%Y-%m-%d')
ipca_expec_ano['Data'] = pd.to_datetime(ipca_expec_ano['Data'], format='%Y-%m-%d')
SELIC_expec_mensais['Data'] = pd.to_datetime(SELIC_expec_mensais['Data'], format='%Y-%m-%d')

# Ajuste da linha de tendência
media_por_dia = ipca_expec_mensais.groupby(ipca_expec_mensais['Data'].dt.date)['Media'].mean().reset_index()

media_por_dia.columns = ['Data', 'Media']

fig_ipca = go.Figure()
fig_ipca.add_trace(go.Scatter(
    x=ipca_expec_mensais['Data'],
    y=ipca_expec_mensais['Media'],
    mode='markers',
    name='IPCA Mensal',
    marker=dict(color='red', size=4)

))
fig_ipca.update_layout(
    title='IPCA Mensal',
    xaxis_title='Data',
    yaxis_title='Média do IPCA',
    template='plotly_white'
)
fig_ipca.show()

fig_ipca_anual = go.Figure()
fig_ipca_anual.add_trace(go.Scatter(
    x=ipca_expec_ano['Data'],
    y=ipca_expec_ano['Media'],
    mode='markers',
    name='IPCA Mensal',
    marker=dict(color='blue', size=4)
))
fig_ipca_anual.update_layout(
    title='IPCA Anual',
    xaxis_title='Data',
    yaxis_title='Média do IPCA',
    template='plotly_white'

)
fig_ipca_anual.show()


print(SELIC_expec_mensais.head())
fig_selic=go.Figure()
fig_selic.add_trace(go.Scatter(
    x=SELIC_expec_mensais['Data'],
    y=SELIC_expec_mensais['Media'],
    mode='markers',
    name='IPCA Mensal',
    #line=dict(color='green', width=3),  # Linha verde com largura personalizada
    marker=dict(color='purple', size=4)  # Ponto roxo com raio de 12
))
fig_selic.update_layout(
    title='Expectativas Selic',
    xaxis_title='Data',
    yaxis_title='Exo. Selic',
    template='plotly_white'
)
fig_selic.show()
```
